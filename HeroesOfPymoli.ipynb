```python
#Dependencies
import pandas as pd
```

```python
#file
csvpath = 'Resources/purchase_data.csv'
```

```python
#create dataframe from CSV file
df = pd.read_csv(csvpath)
df
```

```python
#Player Count
total_players = len(df["SN"].unique())
newdf = pd.DataFrame({"Total Players":[total_players]})
newdf.head()
```

```python
#Purchasing Analysis (total)
total_items = len(df["Item Name"].unique())
average_price = df["Price"].mean()
total_purchases = df["Purchase ID"].count()
total_revenue = df["Price"].sum()

summary = pd.DataFrame([{"Number of Unique Items":total_items,
                        "Average Price": average_price, 
                        "Number of Purchases":total_purchases, 
                        "Total Revenue":total_revenue}])
summary["Average Price"] = summary["Average Price"].map("${:.2f}".format)
summary["Total Revenue"] = summary["Total Revenue"].map("${:,.2f}".format)
summary
```

```python
#Gender Demographics
sn_df = df.drop_duplicates(subset="SN")

total_male = len(sn_df.loc[sn_df["Gender"]=="Male",:])
total_female = len(sn_df.loc[sn_df["Gender"]=="Female",:])
total_other = len(sn_df.loc[sn_df["Gender"]=="Other / Non-Disclosed",:])

male_percent = total_male/total_players*100
female_percent = total_female/total_players*100
other_percent = total_other/total_players*100

gender_df = pd.DataFrame({"Total Count":[total_male, total_female, total_other],
                          "Percentage of Players":[male_percent, female_percent, other_percent],
                          "Gender":["Male", "Female", "Other / Non-Disclosed"]})

gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:.2f}%".format)
gender_df = gender_df.set_index(["Gender"], drop=True)
gender_df
```

```python
#Purchasing Analysis (Age)

purchase_age = df

bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9,sn_df["Age"].max()]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_age["Age Ranges"] = pd.cut(purchase_age["Age"], bins, labels = group_names, include_lowest=False)


purchase_price = purchase_age[["Age Ranges","Price"]]
purchase_price = purchase_price.groupby("Age Ranges")
purchase_price = purchase_price.sum()

purchase_age = purchase_age[["Age Ranges","Age"]]
purchase_age = purchase_age.groupby("Age Ranges")
purchase_age = purchase_age.count()

purchase_summary = pd.merge(purchase_age, purchase_price, left_index=True, right_index=True)
average_price = purchase_summary["Price"]/purchase_summary["Age"]
purchase_summary["Average Purchase Price"] = average_price


purchase_age2 = sn_df


purchase_age2["Age Ranges"] = pd.cut(purchase_age2["Age"], bins, labels = group_names, include_lowest=False)


purchase_age2 = purchase_age2[["Age Ranges","Age"]]
purchase_age2 = purchase_age2.groupby("Age Ranges")
purchase_age2 = purchase_age2.count()


average_price2 = purchase_summary["Price"]/purchase_age2["Age"]


purchase_summary["Average Total Purchase per Person"] = average_price2

purchase_summary = purchase_summary.rename(columns= {"Age":"Purchase Count", "Price":"Total Purchase Value"})
purchase_summary = purchase_summary[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Total Purchase per Person"]]

purchase_summary["Average Purchase Price"] = purchase_summary["Average Purchase Price"].map("${:,.2f}".format)
purchase_summary["Total Purchase Value"] = purchase_summary["Total Purchase Value"].map("${:,.2f}".format)
purchase_summary["Average Total Purchase per Person"] = purchase_summary["Average Total Purchase per Person"].map("${:,.2f}".format)

purchase_summary
```

```python
#Top Spenders
sn_index = df.groupby("SN")

total_spent = sn_index.sum()
total_purchased = sn_index.count()
average_spent = sn_index.mean()

total_spent = total_spent["Price"]
total_purchased = total_purchased["Purchase ID"]
average_spent = average_spent["Price"]

merge1 = pd.merge(total_spent, total_purchased, left_index=True, right_index=True)
merge2 = pd.merge(merge1, average_spent, left_index=True, right_index=True)
merge2 = merge2.rename(columns ={"Price_x":"Total Purchase Value", "Purchase ID":"Purchase Count", "Price_y":"Average Purchase Price"})
merge2 = merge2[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
merge2 = merge2.sort_values("Total Purchase Value",axis=0, ascending=False)
merge2["Average Purchase Price"] = merge2["Average Purchase Price"].map("${:.2f}".format)
merge2["Total Purchase Value"] = merge2["Total Purchase Value"].map("${:.2f}".format)
merge2
```

```python
#Most Popular Items
id_index = df.groupby(["Item ID", "Item Name"])

purchase_count = id_index.count() #purchase id
item_total_price = id_index.sum() #price
item_price = id_index.sum() / id_index.count() #price

purchase_count = purchase_count["Purchase ID"] #purchase id
item_total_price = item_total_price["Price"] #price
item_price = item_price["Price"] #price

merge3 = pd.merge(purchase_count,item_price,left_index = True, right_index = True)
merge4 = pd.merge(merge3, item_total_price,left_index = True, right_index = True)

merge4 = merge4.rename(columns = {"Purchase ID":"Purchase Count", "Price_x":"Item Price", "Price_y":"Total Purchase Value"})
merge5 = merge4.sort_values("Purchase Count", ascending = False)
merge5["Item Price"] =merge5["Item Price"].map("${:.2f}".format)
merge5["Total Purchase Value"] =merge5["Total Purchase Value"].map("${:.2f}".format)
merge5
```

```python
#Most Profitable Items
merge6 = merge4.sort_values("Total Purchase Value", ascending = False)
merge6["Item Price"] =merge6["Item Price"].map("${:.2f}".format)
merge6["Total Purchase Value"] =merge6["Total Purchase Value"].map("${:.2f}".format)
merge6
```
